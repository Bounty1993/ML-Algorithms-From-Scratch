```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
import statsmodels.api as sm
```

```python
X, Y = load_boston().data, load_boston().target
df = pd.DataFrame(X, columns=load_boston().feature_names)
df.head()
```

```python
def forward_selection(df, Y, significant_score):
    columns = list(range(df.shape[1]))
    checked_cols = []
    while True:
        to_check = list(set(columns) - set(checked_cols))
        pvals = []
        print(to_check)
        for col in to_check:
            X_with_bias = sm.add_constant(df[:, checked_cols + to_check])
            model = sm.OLS(Y, X_with_bias).fit()
            pvalue = model.pvalues[col]
            pvals.append(pvalue)
        min_pvalue = min(pvals)
        print(min_pvalue)
        if min_pvalue < significant_score:
            checked_cols.append(np.argmin(pvals))
        else:
            break
        print('Hello')
    return checked_cols
```

```python
columns = list(range(X.shape[1]))
checked_cols = []
to_check = set(columns) - set(checked_cols)
pvals = []
for col in to_check:
    X_with_bias = sm.add_constant(X)
    model = sm.OLS(Y, X_with_bias).fit()
    pvalue = model.pvalues[col]
    pvals.append(pvalue)
min_pvalue = min(pvals)
if min_pvalue < 0.05:
    checked_cols.append(np.argmin(pvals))
```

```python
forward_selection(X, Y, 0.05)
```
