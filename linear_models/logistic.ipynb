```python
def cost(theta, X, y):
    m = X.shape[0]
    h = sigmoid(np.dot(X, theta))
    first = np.dot(np.log(h).T, y)
    second = np.dot(np.log(1-h).T, (1-y))
    J = np.sum(first + second) / m
    grad = np.dot(X.T, (h.ravel() - y)) / m
    return grad
    
```
