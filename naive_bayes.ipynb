```python
import pandas as pd
import numpy as np
from scipy.stats import norm
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
X, y = make_blobs(centers=3)
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
```

```python
def get_priors(y):
    classes, counts = np.unique(y, return_counts=True)
    priors = counts / len(y)
    return list(zip(classes, priors))

def get_distribution(data):
    data_mean = np.mean(data)
    data_std = np.std(data)
    distribution = norm(data_mean, data_std)
    return distribution

def get_class_probability(x, prior, dist_0, dist_1):
    return prior * dist_0.pdf(x[:, 0]) * dist_0.pdf(x[:,1])


first_class = X[y==0]
dist_0 = get_distribution(first_class[:, 0])
dist_1 = get_distribution(first_class[:, 1])
```
